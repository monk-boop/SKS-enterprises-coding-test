```python
import re
import pandas as pd
```

```python
df = pd.read_csv("coupons_ner.csv",header = None)
df.columns = ['coupons']
```

```python
df
```

```python
def value(df):
    values = []
    for i in df.coupons.values:
                                                                    #* - 0 or more
        pattern = re.findall(r'\$\s?\.?\d+\.?[\d+ ]',i)             #+ - 1 or more
        pattern2 = re.findall(r'\d\d[¢]',i)                         #? - 0 or 1
        try:
            values.append(pattern.pop(0))
        except IndexError:
            try:
                values.append(pattern2.pop(0))
            except IndexError:
                values.append('0')
        
    return values
```

```python
x = value(df)
```

```python
x[157]
```

```python
indices = [i for i, j in enumerate(x) if j == '0']
```

```python
print(df['coupons'].values[157])
```
